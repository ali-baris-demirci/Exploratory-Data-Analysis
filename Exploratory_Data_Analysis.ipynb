```python
import pandas as pd
```

```python
xlsx = pd.ExcelFile('Total Data.xlsx')
customers = pd.read_excel(xlsx, 'Customers')
products = pd.read_excel(xlsx, 'Products')
sales = pd.read_excel(xlsx, 'Sales')
territory = pd.read_excel(xlsx, 'Territory')
budget = pd.read_excel(xlsx, 'Budget')
calendar = pd.read_excel(xlsx, 'Calender')
```

```python
customers.info()
print('-----------------------------------------------------------------')
products.info()
print('-----------------------------------------------------------------')
sales.info()
print('-----------------------------------------------------------------')
territory.info()
print('-----------------------------------------------------------------')
budget.info()
print('-----------------------------------------------------------------')
calendar.info()
```

```python
budget = pd.melt(budget, id_vars=['Category', 'Subcategory', 'ProductName', 'ProductKey'], var_name='MonthYear', value_name='Budget')
budget
```

```python
calendar['MonthYear'] = calendar['Month'] + ', ' + calendar['Year'].astype(str)
calendar = calendar[calendar['Year'] != 2017]
calendar
```

```python
customers.reset_index(drop=True, inplace=True)
products.reset_index(drop=True, inplace=True)
sales.reset_index(drop=True, inplace=True)
territory.reset_index(drop=True, inplace=True)
sales.reset_index(drop=True, inplace=True)
calendar.reset_index(drop=True, inplace=True)
```

```python
writer = pd.ExcelWriter('Total Data - Unpivoted.xlsx')
customers.to_excel(writer, sheet_name = 'Customers', index=False)
products.to_excel(writer, sheet_name = 'Products', index=False)
sales.to_excel(writer, sheet_name = 'Sales', index=False)
territory.to_excel(writer, sheet_name = 'Territory', index=False)
budget.to_excel(writer, sheet_name = 'Budget', index=False)
calendar.to_excel(writer, sheet_name = 'Calendar', index=False)
writer.close()
```
